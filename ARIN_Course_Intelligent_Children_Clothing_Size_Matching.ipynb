```python
# Import libraries
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

**Brand A**

**========================================**

```python
# Generate categories - Height_Brand A
brA_height = ctrl.Antecedent(np.arange(98.60, 155.0, 0.01), 'brA_height')

# Then, use them to set ranges for the fuzzy variable
brA_height['s1']  = fuzz.trimf(brA_height.universe, [98.60, 111.94, 116.0])
brA_height['s2']  = fuzz.trimf(brA_height.universe, [98.60, 118.70, 123.0])
brA_height['s3'] = fuzz.trimf(brA_height.universe, [98.60, 123.52, 128.0])
brA_height['s4']  = fuzz.trimf(brA_height.universe, [98.60, 129.31, 134.0])
brA_height['s5']  = fuzz.trimf(brA_height.universe, [98.60, 133.17, 138.0])
brA_height['s6'] = fuzz.trimf(brA_height.universe, [98.60, 137.03, 142.0])
brA_height['s7'] = fuzz.trimf(brA_height.universe, [98.60, 143.79, 149.0])
brA_height['s8'] = fuzz.trimf(brA_height.universe, [98.60, 149.58, 155.0])

# Visualization
brA_height.view()
```

```python
# Generate categories - Weight_Brand A
brA_weight = ctrl.Antecedent(np.arange(17.0, 45.0, 0.01), 'brA_weight')

# Then, use them to set ranges for the fuzzy variable
brA_weight['s1']  = fuzz.trimf(brA_weight.universe, [17.0, 19.30, 20.00])
brA_weight['s2']  = fuzz.trimf(brA_weight.universe, [17.0, 21.23, 22.00])
brA_weight['s3'] = fuzz.trimf(brA_weight.universe, [17.0, 22.20, 23.00])
brA_weight['s4']  = fuzz.trimf(brA_weight.universe, [17.0, 24.13, 25.0])
brA_weight['s5']  = fuzz.trimf(brA_weight.universe, [17.0, 27.02, 28.0])
brA_weight['s6'] = fuzz.trimf(brA_weight.universe, [17.0, 31.85, 33.0])
brA_weight['s7'] = fuzz.trimf(brA_weight.universe, [17.0, 37.64, 39.0])
brA_weight['s8'] = fuzz.trimf(brA_weight.universe, [17.0, 43.43, 45.0])

# Visualization
brA_weight.view()
```

```python
# Generate categories - Waist_Brand A
brA_waist = ctrl.Antecedent(np.arange(45.05, 66.0, 0.01), 'brA_waist')

# Then, use them to set ranges for the fuzzy variable
brA_waist['s1']  = fuzz.trimf(brA_waist.universe, [45.05, 51.15, 53.00])
brA_waist['s2']  = fuzz.trimf(brA_waist.universe, [45.05, 53.08, 55.00])
brA_waist['s3'] = fuzz.trimf(brA_waist.universe, [45.05, 54.04, 56.00])
brA_waist['s4']  = fuzz.trimf(brA_waist.universe, [45.05, 55.01, 57.0])
brA_waist['s5']  = fuzz.trimf(brA_waist.universe, [45.05, 55.97, 58.0])
brA_waist['s6'] = fuzz.trimf(brA_waist.universe, [45.05, 58.87, 61.0])
brA_waist['s7'] = fuzz.trimf(brA_waist.universe, [45.05, 61.76, 64.0])
brA_waist['s8'] = fuzz.trimf(brA_waist.universe, [45.05, 63.69, 66.0])

# Visualization
brA_waist.view()
```

```python
# Generate data - Size - Brand A
brA_size = ctrl.Consequent(np.arange(4.0, 15.0, 0.01), 'brA_size')

# Then, use them to set ranges for the fuzzy variable
brA_size['5']  = fuzz.trimf(brA_size.universe, [4.5, 5.0, 5.5])
brA_size['6']  = fuzz.trimf(brA_size.universe, [5.5, 6.0, 6.5])
brA_size['6X'] = fuzz.trimf(brA_size.universe, [5.5, 6.5, 7.0])
brA_size['7']  = fuzz.trimf(brA_size.universe, [6.5, 7.0, 8.5])
brA_size['8']  = fuzz.trimf(brA_size.universe, [7.5, 8.0, 9.5])
brA_size['10'] = fuzz.trimf(brA_size.universe, [9.5, 10.0, 11.5])
brA_size['12'] = fuzz.trimf(brA_size.universe, [11.5, 12.0, 13.5])
brA_size['14'] = fuzz.trimf(brA_size.universe, [13.0, 14.0, 14.5])

# Visualization
brA_size.view()
```

```python
# Rules
rule_1 = ctrl.Rule((brA_height['s1']&brA_weight['s1'])|
                   (brA_height['s1']&brA_waist['s1'])|
                   (brA_weight['s1']&brA_waist['s1'])
                   , brA_size['5'], '5')

rule_2 = ctrl.Rule((brA_height['s2']&brA_weight['s2'])|
                   (brA_height['s2']&brA_waist['s2'])|
                   (brA_weight['s2']&brA_waist['s2'])
                   , brA_size['6'], '6')

rule_3 = ctrl.Rule((brA_height['s3']&brA_weight['s3'])|
                   (brA_height['s3']&brA_waist['s3'])|
                   (brA_weight['s3']&brA_waist['s3'])
                   , brA_size['6X'], '6X')

rule_4 = ctrl.Rule((brA_height['s4']&brA_weight['s4'])|
                   (brA_height['s4']&brA_waist['s4'])|
                   (brA_weight['s4']&brA_waist['s4'])
                   , brA_size['7'], '7')

rule_5 = ctrl.Rule((brA_height['s5']&brA_weight['s5'])|
                   (brA_height['s5']&brA_waist['s5'])|
                   (brA_weight['s5']&brA_waist['s5'])
                   , brA_size['8'], '8')

rule_6 = ctrl.Rule((brA_height['s6']&brA_weight['s6'])|
                   (brA_height['s6']&brA_waist['s6'])|
                   (brA_weight['s6']&brA_waist['s6'])
                   , brA_size['10'], '10')

rule_7 = ctrl.Rule((brA_height['s7']&brA_weight['s7'])|
                   (brA_height['s7']&brA_waist['s7'])|
                   (brA_weight['s7']&brA_waist['s7'])
                   , brA_size['12'], '12')

rule_8 = ctrl.Rule((brA_height['s8']&brA_weight['s8'])|
                   (brA_height['s8']&brA_waist['s8'])|
                   (brA_weight['s8']&brA_waist['s8'])
                   , brA_size['14'], '14')
```

```python
# Train and Create a predictor
size_brA_predictor = ctrl.ControlSystem([rule_1, rule_2, rule_3, rule_4,
                                         rule_5, rule_6, rule_7, rule_8,
                                        ])
brA_pred = ctrl.ControlSystemSimulation(size_brA_predictor)
```

```python
# Test an instance - Darwish (From Ref. Source)
brA_pred.input['brA_height'] = 125
brA_pred.input['brA_weight'] = 22.1
brA_pred.input['brA_waist']  = 55
brA_pred.compute()

# Print
size_pred = brA_pred.output['brA_size']
dis_str = "Predicted: {size}".format(size = size_pred)   # Print predicted value
print(dis_str)
brA_size.view(sim = brA_pred)   # Display on graph
```
